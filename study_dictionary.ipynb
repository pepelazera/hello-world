Here, we have a standard mode to use dictionaries in python


```python
student = {
    'name': 'Pedro',
    'age': 22,
    'course': 'Software engineering'
}

print(f"{student['name'], student['age'], student['course']}")
```

And here, I create a dictionary with a list and a tuple inside of it

```python
numbs = [('a', 1), ('b', 2), ('c', 3)]
dict_1 = dict(numbs)

print(dict_1) # {'a': 1, 'b': 2, 'c': 3}
```

Now, I will use the get() command

```python
student = {
    'name': 'Pedro',
    'age': 22,
    'course': 'Software engineering'
}

name = student['name']
age = student.get('age') # 22
email = student.get('email', 'not registered')

print(f"Name: {name} -- Age: {age} -- Email: {email}")
```

And now, I will verify if keys are exist

```python
student = {
    'name': 'Pedro',
    'age': 22,
    'course': 'Software engineering'
}

if 'course' in student:
    print(f"Course: {student['course']}")
```

```python
text = 'python is a programming language'
words = text.split()
count_words = {}

for word in words:
    cc = 0
    cc += len(count_words)
    count_words[word] = count_words.get(word, cc+1)
    print(f"{count_words}")
```
